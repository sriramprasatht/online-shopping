```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Firefox()
driver.get("https://www.flipkart.com/laptops/asus~brand/pr?sid=6bo,b5g")
```

```python
html_content = driver.page_source
```

```python
html_content
```

```python
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
soup
```

```python
df=soup.find_all('div',class_="tUxRFH")
```

```python
df
```

```python
len(df)
```

```python
lp_names=[]
price=[]
discount=[]
for i in df:
    laptop=i.find('div',class_="KzDlHZ")
    lp_names.append(laptop.get_text())
    p=i.find('div',class_="Nx9bqj _4b5DiR")
    price.append(p.text)
    d=i.find('div',class_="UkUFwK")
    discount.append(d.string)
   

print(lp_names)
print(price)
print(discount)
```

```python
import pandas as pd
df=pd.DataFrame({'laptop_name':lp_names,'price':price,'discount':discount})
```
